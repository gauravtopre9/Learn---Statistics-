#  Types of Sampling

* Random Sampling : Each element of the population is having equal probabiliy of getting picked in sample set.


* systematic Sampling : Every Nth element is getting picked in the sample set.


* Stratified Sampling : Sample sets are picked in such a way that they do not overlap.


* Convenience Sampling : Domain based sampling.

```python
import pandas as pd
  
# Create a dictionary of students
df= pd.DataFrame({
    'Name': ['Lisa', 'Kate', 'Ben', 'Kim', 'Josh',
             'Alex', 'Evan', 'Greg', 'Sam', 'Ella'],
    'ID': ['001', '002', '003', '004', '005', '006', 
           '007', '008', '009', '010'],
    'Grade': ['A', 'A', 'C', 'B', 'B', 'B', 'C', 
              'A', 'A', 'A'],
    
    'Category': [2, 3, 1, 3, 2, 3, 3, 1, 2, 1]
})
  
```

**Disproportionate Sampling:** 

Using pandas groupby, separate the students into groups based on their grade i.e A, B, C and randomly sample 2 students from each grade group using the sample function

```python
df.groupby('Grade', group_keys=False).apply(lambda x: x.sample(2))
```

**Proportionate Sampling:** 

Using pandas groupby, separate the students into groups based on their grade i.e A, B, C, and random sample from each group based on population proportion. The total sample size is 60%(0.6) of the population

```python
df.groupby('Grade', group_keys=False).apply(lambda x: x.sample(frac=0.6))
```
